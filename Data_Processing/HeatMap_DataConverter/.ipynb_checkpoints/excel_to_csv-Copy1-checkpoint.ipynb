## Balance Sheet CSV

```python
import pandas as pd
```

```python
tracked_country_name_list = ["United Arab Emirates", "Austria", "Australia", "Belgium", "Canada", "Switzerland", "China", "Czech Republic", "Germany", "Denmark", "Egypt", "Spain", "Finland", "France", "United Kingdom", "Greece", "Hong Kong", "Hungary", "Indonesia", "Ireland", "Israel", "India", "Italy", "Jordan", "Korea", "Luxembourg", "Malaysia", "Netherlands", "Norway", "New Zealand", "Philippines", "Poland", "Portugal", "Qatar", "Russia", "Saudi Arabia", "Sweden", "Singapore", "Thailand", "Turkey", "Taiwan", "United States", "South Africa"]
```

```python
bs_df = pd.read_excel('Valuation_data_cleaned.xlsx', sheet_name = 'Balance Sheet', header =24, nrows=44)

bs_global_intercept = bs_df.iloc[0,2]
# print(bs_global_intercept)

# remove global intercept row
bs_df = bs_df.iloc[1:, :]

# select only ISO and Coef column
bs_df = bs_df[["Variable", "Coef"]]

# rename column names to id and value to match js script
bs_df = bs_df.rename(columns={'Variable':"id", "Coef":"country coefficient"})

# df = df.set_index("id")
# df = df.set_index(range(df.shape[0]))
# df = df.reset_index()

# country scaling factor
bs_df["value"] = bs_df['country coefficient'] + bs_global_intercept

# round values to 4 decimal places
bs_df = bs_df.round({'country coefficient': 4, 'value': 4})

bs_df["name"] = tracked_country_name_list

# print(df.shape[0])
# print(len(tracked_country_name_list))

# rearrange order of columns
bs_df = bs_df[["id","name", "country coefficient", "value"]]
bs_df

```

```python
bs_csv_file = bs_df.to_csv(r'balance_sheet.csv')
```

```python

bs_df
# print(df.to_string())

for i in range(len(bs_df)):
    bs_string = "{ " + '''id: "{}", value: {}, "url": "../Compiled/Select_Industry/Select Industry_{}.html"'''.format(bs_df.iloc[i, 0], bs_df.iloc[i, 3], bs_df.iloc[i, 1]) + " },"
    print(bs_string)
    
```

## Income Statement CSV

```python
is_df = pd.read_excel("Valuation_data_cleaned.xlsx", sheet_name = "Income Statement", header = 25, nrows = 44)

is_global_average = is_df.iloc[0,2]
# print(is_global_average)

# remove global intercept row
is_df = is_df[1:]

is_df = is_df[["Variable", "Coef"]]
is_df = is_df.rename(columns={"Variable":"id", "Coef":"country coefficient"})

is_df["name"] = tracked_country_name_list

is_df["value"] = is_df["country coefficient"] + is_global_average

is_df = is_df.round({"country coefficient": 4, "value": 4})

is_df = is_df[["id", "name", "country coefficient", "value"]]

is_df
```

```python
is_csv_file = is_df.to_csv(r'income_statement.csv')
```
