## Balance Sheet CSV

```python
import pandas as pd
```

```python
df = pd.read_excel('Valuation_data_cleaned.xlsx', sheet_name = 'Balance Sheet', header =24, nrows=43)
# remove global average row
df = df[1:]

# select only ISO and Coef column
df = df[["Variable", "Coef"]]

# rename column names to id and value to match js script
df = df.rename(columns={'Variable':"id", "Coef":"value"})

df = df.set_index('id')
df
```

```python
csv_file = df.to_csv(r'balance_sheet.csv')
```

```python
df = pd.DataFrame(df)
print(df.to_string())
```

## Income Statement CSV

```python
df1 = pd.read_excel("Valuation_data_cleaned.xlsx", sheet_name = "Income Statement", header = 25, nrows = 44)

df1 = df1[1:]
df1 = df1[["Variable", "Coef"]]
df1 = df1.rename(columns={"Variable":"id", "Coef":"value"})
df1 = df1.set_index("id")

df1
```

```python
csv_file = df1.to_csv(r'income_statement.csv')
```

```python
df1 = pd.DataFrame(df1)
print(df1.to_string())
```
